```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
df = pd.read_csv(io.BytesIO(uploaded['airport-all.csv']))
```

```python
df
```

```python
df = df.replace(to_replace='None', value=np.nan)
```

```python
df.isnull().sum()
```

```python
df = df.drop(['To Date','From Date'], axis=1)
```

```python
df
```

```python
col = df.pop('Date')
df.insert(0, 'Date', col)
```

```python
df
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
```

```python
df.dtypes
```

```python
df[['PM2.5','PM10','NO','NO2','NOx','NH3','SO2','CO','Ozone','Benzene','Toluene','Eth-Benzene','MP-Xylene','RH','WS','WD','SR','BP','Xylene','AT']] = df[['PM2.5','PM10','NO','NO2','NOx','NH3','SO2','CO','Ozone','Benzene','Toluene','Eth-Benzene','MP-Xylene','RH','WS','WD','SR','BP','Xylene','AT']].astype(float)
```

```python
pm2_df = df[["Date", "PM2.5"]]
```

```python
pm2_df
```

```python
pm2_df = pm2_df.dropna()
```

```python
pm2_df['Date'].dt.weekofyear
```

```python
pm2_df['week_number'] = pm2_df['Date'].dt.isocalendar().week
```

```python
pm2_df.groupby('week_number')['PM2.5'].mean()
```

```python
pm2_df
```

```python
pm2_df['Year'] = pm2_df['Date'].dt.isocalendar().year
```

```python
pm2_df
```

```python
pm2_df.groupby('week_number')['PM2.5'].mean()
```

```python
pm2_df = pm2_df.groupby(['Year', 'week_number'], as_index=False)['PM2.5'].mean()
```

```python
pm2_df.groupby(['week_number','Year'])['PM2.5'].mean()
```

```python
pm2_df
```

```python
# conditions = [(pm2_df['Year'] == 2019),
#               (pm2_df['Year'] == 2020),
#               (pm2_df['Year'] == 2021),
#               (pm2_df['Year'] == 2022),
#               (pm2_df['Year'] == 2023)]
# week_no = [pm2_df['week_number'].apply(lambda x: x),
#            pm2_df['week_number'].apply(lambda x: x+52),
#            pm2_df['week_number'].apply(lambda x: x+105),
#            pm2_df['week_number'].apply(lambda x: x+157),
#            pm2_df['week_number'].apply(lambda x: x+209)]
# pm2_df['week_no'] = np.select(conditions, week_no)
pm2_df.loc[pm2_df['Year'] == 2019, 'week_number'] += 0
pm2_df.loc[pm2_df['Year'] == 2020, 'week_number'] += 52
pm2_df.loc[pm2_df['Year'] == 2021, 'week_number'] += 105
pm2_df.loc[pm2_df['Year'] == 2022, 'week_number'] += 157
pm2_df.loc[pm2_df['Year'] == 2023, 'week_number'] += 209
```
